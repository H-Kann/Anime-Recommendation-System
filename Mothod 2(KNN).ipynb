```python
# importing the libraries
import numpy as np
import pandas as pd

# Loading anime.csv data
missing_values=["n/a","na","--","?"]
df=pd.read_csv("anime.csv",na_values= missing_values,delimiter=',')
df
```

```python
# Dropping missing values
df.dropna(inplace=True)
```

```python
# Loading rating.csv data
rating_df = pd.read_csv('rating.csv')
rating_df.head()
```

```python
rating_df.shape
```

```python
# Merging the two datasets
df_merge = pd.merge(df, rating_df, on = 'anime_id')
df_merge.head()
```

```python
df_merge.shape
```

```python
# Taking out the rows with no ratings, that are represented as -1. 
df_merge.rating_x.min()
```

```python
df_merge = df_merge[df_merge.rating_y != -1]

# Showing the result
df_merge.head()
```

```python
df_merge.shape
```

```python
# Counting the number of total votes per anime
temp = df_merge.groupby(['anime_id'], as_index=False).agg(totalvotes=('anime_id', pd.Series.count))
```

```python
temp.head()
```

```python
# Cheking the result
temp.loc[temp['anime_id'] == 32281].head()
```

```python
# Creating a final dataset
finaldataset = pd.merge(df_merge, temp, on = 'anime_id')
finaldataset.head()
```

```python
# Removing all the duplicate rows and leaving each anime with one row
finaldataset.drop_duplicates(subset=['anime_id'], inplace=True)
finaldataset.head(20)
```

<h2>Applying the K nearest neighbor</h2>

```python
# As the genre and type variables are categorical data, we need to one-hot encode them to ensure we can match incoming instances on quantitative similarity
from sklearn.preprocessing import LabelEncoder

enconder = LabelEncoder()
genre_labels = enconder.fit_transform(finaldataset['genre'])
genre_mappings = {index: label for index, label in 
                 enumerate(enconder.classes_)}

# Creating a map for the name column
enconder = LabelEncoder()
name_labels = enconder.fit_transform(finaldataset['name'])
name_mappings = {index: label for index, label in 
                  enumerate(enconder.classes_)}
```

```python
# merging encoded data back to dataframe and visualise which columns they exist in
finaldataset['genre_mappings'] = genre_labels

finaldataset['name_mappings'] = name_labels


# Converting the dataset to a csv file and saving it locally
finaldataset.to_csv('finaldataset.csv', index=False)


# Showing the result
X =finaldataset.loc[:,["name_mappings", "genre_mappings"]]
X[0:5]
```

```python
# isolating an anime that we want to match
test2= finaldataset.loc[finaldataset['name'] == "Hajime no Ippo"]
test2
```

```python
# Applying the K-nearest neighbor
from sklearn.neighbors import NearestNeighbors

# Applying the name_mapping value and working with the seven nearest values
rec=[3004,1720]
Knn=NearestNeighbors(n_neighbors=7).fit(X)
print(Knn.kneighbors([rec])[1][0])
```
